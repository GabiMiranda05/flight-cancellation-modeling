```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.metrics import classification_report, accuracy_score
```

```python
df = pd.read_csv("/content/drive/MyDrive/Datasets/dataset_merged_processed_no_outliers_numeric.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
X = df.drop(columns=["YEAR","CANCELLATION_REASON","AIRLINE_NAME","ORIGIN_LATITUDE","ORIGIN_LONGITUDE","DESTINATION_LONGITUDE" ,"DESTINATION_LATITUDE","DIVERTED", "CANCELLED","CANCELLED_NUM","ORIGIN_AIRPORT","DESTINATION_AIRPORT","ORIGIN_STATE","DESTINATION_STATE"])
y = df["CANCELLED_NUM"]
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
decision_tree = DecisionTreeClassifier(max_depth=8, min_samples_leaf=50, class_weight= 'balanced', criterion = 'entropy' , random_state=42)
```

```python
decision_tree.fit(X_train, y_train)
```

```python
y_pred_train_dt = decision_tree.predict(X_train)
y_pred_test_dt = decision_tree.predict(X_test)
```

```python
y_pred_train_dt
```

```python
print("train")
print(classification_report(y_train,y_pred_train_dt))
print("------------------------------------------------------")
print("test")
print(classification_report(y_test,y_pred_test_dt))
```

Dado el fuerte desbalance de clases, se ajustaron los hiperparÃ¡metros del Ã¡rbol de decisiÃ³n y se incorporÃ³ ponderaciÃ³n de clases para priorizar la detecciÃ³n de vuelos cancelados. Esto permitiÃ³ incrementar el recall de la clase minoritaria del 20% al 77%, a costa de una disminuciÃ³n en la precisiÃ³n y en la exactitud global.

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
random_forest = RandomForestClassifier(n_estimators=300, max_depth=12, min_samples_leaf=50, class_weight="balanced",n_jobs=-1, random_state=42)
```

```python
random_forest.fit(X_train, y_train)
```

```python
y_pred_train_rf = random_forest.predict(X_train)
y_pred_test_rf = random_forest.predict(X_test)
```

```python
print("train")
print(classification_report(y_train,y_pred_train_rf))
print("------------------------------------------------------")
print("test")
print(classification_report(y_test,y_pred_test_rf))
```

El modelo prioriza la detecciÃ³n de vuelos cancelados (alto recall), aceptando un mayor nÃºmero de falsos positivos, lo cual es adecuado en un contexto donde anticipar cancelaciones es mÃ¡s crÃ­tico que predecirlas errÃ³neamente.

No se utilizÃ³ KNN en este proyecto porque, aunque es un modelo sencillo, no escala bien a conjuntos de datos grandes y desbalanceados como el utilizado (mÃ¡s de 600.000 registros). KNN presenta tiempos de predicciÃ³n muy elevados, ya que debe calcular distancias contra todos los puntos de entrenamiento, lo que lo vuelve poco viable computacionalmente.

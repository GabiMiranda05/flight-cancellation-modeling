```python
df["CANCELLED"] = df["CANCELLED"].replace({"No Cancelado":0, "Cancelado":1})
```

```python
df["MONTH"] = df["MONTH"].replace({"1":0, "2":1})
```

```python
region_map = {
    'ME': 'Noreste', 'NH': 'Noreste', 'VT': 'Noreste', 'MA': 'Noreste', 'RI': 'Noreste',
    'CT': 'Noreste', 'NY': 'Noreste', 'NJ': 'Noreste', 'PA': 'Noreste',

    'OH': 'MedioOeste', 'MI': 'MedioOeste', 'IN': 'MedioOeste', 'IL': 'MedioOeste', 'WI': 'MedioOeste',
    'MN': 'MedioOeste', 'IA': 'MedioOeste', 'MO': 'MedioOeste', 'ND': 'MedioOeste', 'SD': 'MedioOeste',
    'NE': 'MedioOeste', 'KS': 'MedioOeste',

    'DE': 'Sur', 'MD': 'Sur', 'DC': 'Sur', 'VA': 'Sur', 'WV': 'Sur', 'KY': 'Sur', 'TN': 'Sur',
    'NC': 'Sur', 'SC': 'Sur', 'GA': 'Sur', 'FL': 'Sur', 'AL': 'Sur', 'MS': 'Sur', 'AR': 'Sur',
    'LA': 'Sur', 'OK': 'Sur', 'TX': 'Sur',

    'MT': 'Oeste', 'ID': 'Oeste', 'WY': 'Oeste', 'CO': 'Oeste', 'NM': 'Oeste', 'AZ': 'Oeste', 'UT': 'Oeste',
    'NV': 'Oeste', 'WA': 'Oeste', 'OR': 'Oeste', 'CA': 'Oeste', 'HI': 'Oeste', 'AK': 'Oeste',

    'PR': 'Territorios', 'VI': 'Territorios', 'GU': 'Territorios', 'AS': 'Territorios'
}


df['ORIGIN_REGION'] = df['ORIGIN_STATE'].map(region_map)
df['DESTINATION_REGION'] = df['DESTINATION_STATE'].map(region_map)
```

```python
df.ORIGIN_REGION.unique()
```

```python
df.DESTINATION_REGION.unique()
```

```python
top_origin = df['ORIGIN_AIRPORT'].value_counts().nlargest(20).index
df['ORIGIN_AIRPORT_REDUCED'] = df['ORIGIN_AIRPORT'].where(df['ORIGIN_AIRPORT'].isin(top_origin), 'OTROS')


top_dest = df['DESTINATION_AIRPORT'].value_counts().nlargest(20).index
df['DESTINATION_AIRPORT_REDUCED'] = df['DESTINATION_AIRPORT'].where(df['DESTINATION_AIRPORT'].isin(top_dest), 'OTROS')
```

```python
columnas_cat = ['ORIGIN_AIRPORT_REDUCED', 'DESTINATION_AIRPORT_REDUCED',
                'ORIGIN_REGION', 'DESTINATION_REGION']
```

```python
ohe = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
```

```python
ohe.fit(df[columnas_cat])
```

```python
df_encoded = ohe.transform(df[columnas_cat])
```

```python
columnas_ohe = ohe.get_feature_names_out(columnas_cat)
df_ohe = pd.DataFrame(df_encoded, columns=columnas_ohe, index=df.index)
```

```python
df_final = pd.concat([df.drop(columns=columnas_cat), df_ohe], axis=1)
```

```python
df_final.to_csv("dataset_merged_processed_no_outliers_numeric.csv", index=False)
```

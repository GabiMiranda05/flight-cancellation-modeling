```python
cols_postvuelo = [
    'DEPARTURE_TIME', 'DEPARTURE_DELAY', 'TAXI_OUT', 'WHEELS_OFF',
    'ELAPSED_TIME', 'AIR_TIME', 'WHEELS_ON', 'TAXI_IN', 'ARRIVAL_TIME',
    'ARRIVAL_DELAY',
    'AIR_SYSTEM_DELAY', 'SECURITY_DELAY', 'AIRLINE_DELAY',
    'LATE_AIRCRAFT_DELAY', 'WEATHER_DELAY'
]
```

```python
cols_irrelevantes = [
    'FLIGHT_NUMBER', 'TAIL_NUMBER',
    'ORIGIN_AIRPORT_NAME', 'DESTINATION_AIRPORT_NAME',
    'ORIGIN_CITY', 'DESTINATION_CITY',
    'ORIGIN_COUNTRY', 'DESTINATION_COUNTRY',
]
```

```python
cols_a_eliminar = cols_postvuelo + cols_irrelevantes
df= df.drop(columns=cols_a_eliminar)
```

```python
df['CANCELLATION_REASON'].unique()
```

```python
reason_map = {
    'A': 'Airline',
    'B': 'Weather',
    'C': 'National Air System',
    'D': 'Security'
}

df['CANCELLATION_REASON'] = df['CANCELLATION_REASON'].map(reason_map)
df['CANCELLATION_REASON'] = df['CANCELLATION_REASON'].fillna('No Aplica')
```

```python
df['CANCELLED'] = df['CANCELLED'].map({0: 'No Cancelado', 1: 'Cancelado'})
```

```python
df = df.fillna(df.mean(numeric_only=True))
```

```python
df['ORIGIN_STATE'].value_counts(normalize=True).head()

valid_states = df['ORIGIN_STATE'].dropna().unique()
state_proportions = df['ORIGIN_STATE'].value_counts(normalize=True)

num_nulls_origin = df['ORIGIN_STATE'].isnull().sum()
num_nulls_dest = df['DESTINATION_STATE'].isnull().sum()
```

```python
random_origin_states = np.random.choice(
    state_proportions.index,
    size=num_nulls_origin,
    p=state_proportions.values
)

random_destination_states = np.random.choice(
    state_proportions.index,
    size=num_nulls_dest,
    p=state_proportions.values
)

df.loc[df['ORIGIN_STATE'].isnull(), 'ORIGIN_STATE'] = random_origin_states
df.loc[df['DESTINATION_STATE'].isnull(), 'DESTINATION_STATE'] = random_destination_states
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
prob_airlines = df['AIRLINE_NAME'].value_counts(normalize=True)

random_airlines = np.random.choice(
    prob_airlines.index,
    size=df['AIRLINE_NAME'].isnull().sum(),
    p=prob_airlines.values
)

df.loc[df['AIRLINE_NAME'].isnull(), 'AIRLINE_NAME'] = random_airlines
```

```python
df['DESTINATION_LATITUDE'] = df['DESTINATION_LATITUDE'].fillna(
    df.groupby('DESTINATION_AIRPORT')['DESTINATION_LATITUDE'].transform('median')
)

df['DESTINATION_LONGITUDE'] = df['DESTINATION_LONGITUDE'].fillna(
    df.groupby('DESTINATION_AIRPORT')['DESTINATION_LONGITUDE'].transform('median')
)
```
